```python
import pandas as pd
import csv


def excel():
    f=pd.read("diabetes_prediction_dataset.csv")
    print(f)

tabela = []

def ler():
    f = open("diabetes_prediction_datset.csv", mode = "r")
    inf = csv.reader(f,delimiter=",")
    nlinha  = 0
    for linha in inf:
        if nlinha == 0:
            nlinha = nlinha + 1
        else:
            tabela.append((linha[0],float(linha[1]),int(linha[2]),int(linha[3]),linha[4],float(linha[5]),float(linha[6]),int(linha[7]),int(linha[8])))

def genero():
    ler()
    contador_male = 0
    contador_female = 0
    for t in tabela:
        genero,_,_,doenca_cardiaca,_,_,_,_,_ = t
        if doenca_cardiaca == 1 and genero == "Female":
            contador_female = contador_female + 1
        elif doenca_cardiaca == 1 and genero == "Male":
            contador_male = contador_male + 1
    
    print("Contador feminino:", contador_female)
    print("Contador masculino:", contador_male)

def distribuicao_etaria():
    ler()
    escaloes = [[0-10],[11-20],[21-30],[41-50],[51-60],[61-70],[71-80]]
    lista=[]
    for t in tabela:
        _,idade,_,doenca_cardiaca,_,_,_,_,_=t
        for escalao in escaloes:
            if doenca_cardiaca==1 and idade>=escalao[0] and idade <=escalao[1]:
                lista.append(escalao)
    for escalao in escaloes:
        print("Escalão ",escalao," tem:",lista.count(escalao)," pessoas")

def distribuicao_glucose():
    ler()
    lista=[]
    niveis_glucose=[]
    for i in range(0, 340,10):
            niveis_glucose.append([i,i+10])
    for t in tabela:
        _,_,_,doenca_cardiaca,_,_,_,nivel_glucose,_=t
        for nivel in niveis_glucose:
            if doenca_cardiaca==1 and nivel_glucose>=nivel[0] and nivel_glucose <=nivel[1]:
                lista.append(nivel)
    for nivel in niveis_glucose:
        print("Nivel", nivel, "tem:", lista.count(nivel), "niveis")



def tabelar():
    num=int(input("Escolha a distribuição que pretende tabelar:\n (1)sexo \n (2)idade \n (3)glucose"))
    if num==1:
        ler()
        contador_female=0
        contador_male=0
        for t in tabela:
            sexo,_,_,doenca_cardiaca,_,_,_,_,_=t
            if doenca_cardiaca==1 and sexo== "Male":
                contador_male = contador_male + 1
            elif doenca_cardiaca==1 and sexo=="Female":
                contador_female = contador_female + 1
        print("Sexo masculino       |    Sexo Feminino        ")
        print("   ",contador_male,"                  ", contador_female)
    elif num==2:
        ler()
        escalões=[[20,24],[25,29],[30,34],[35,39],[40,44],[45,49],[50,54],[55,59],[60,64],[65,69],[70,74],[75,79],[80,84],[85,89]]
        lista=[]
        for t in tabela:
            _,idade,_,doenca_cardiaca,_,_,_,_,_=t
            for escalao in escalões:
                if doenca_cardiaca==1 and idade>=escalao[0] and idade <=escalao[1]:
                    lista.append(escalao)
        print(" Escalão      |   Idade    ")
        for escalao in escalões:
            print("   ",escalao,"   ", idade)
    else:
        ler()
        lista=[]
        niveis_glucose=[]
        for i in range(0, 340,10):
                niveis_glucose.append([i,i+10])
        for t in tabela:
            _,_,_,doenca_cardiaca,_,_,_,nivel_glucose,_=t
            for nivel in niveis_glucose:
                if doenca_cardiaca==1 and nivel_glucose>=nivel[0] and nivel_glucose <=nivel[1]:
                    lista.append(nivel)
        print(" Nivel  |  Pessoas")
        for nivel in niveis_glucose:
            print ( nivel, "  ",lista.count(nivel))


def menu():
    opcao= int(input("introduza 1 nº entre 0 e 5: 0-TabelaExcel(); 1-lerinformação(); 2-distribuiçãoSexo(); 3-distribuiçãoEtaria(); 4-DistribuiçãoGlucose(); 5-Tabelar()"))
    if opcao==0:
        excel()
    elif opcao==1:
        ler()
    elif opcao==2:
       genero()
    elif opcao ==3:
        distribuicao_etaria()
    elif opcao==4:
        distribuicao_glucose()
    elif opcao==5:
        tabelar()

menu()

#f.close()   #no final fechar sempre
```
