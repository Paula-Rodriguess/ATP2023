```python
#TPC3

import random

n = int(input("Escolha 1 se quer adivinhar o número ou 2 para o computador adivinhar!!"))
if n ==1 :
    numero = random.randint(0,101)
    contador = 1
    numero_escolhido = int(input("Adivinhe o número"))
    while numero != numero_escolhido:
        contador = contador + 1
        if numero_escolhido < numero:
            print("O número correto é maior!")
        elif numero_escolhido > numero:
            print("O número correto é menor!")
        

        
        numero_escolhido = int(input("Escolha outro número"))
    print("Acertou o número em", contador, "tentativas!!!")

elif n == 2:
    numero_selecionado = int(input("Escolha um valor entre 0 e 100."))
    limSup = 100
    limInf = 0
    escolhaComp = 50
    tentativas = 1
    while escolhaComp != numero_selecionado:
        print("O número que escolhi foi {}, é maior ou menor do que o número escolhido?")
        if escolhaComp > numero_selecionado:
            limSup = escolhaComp
            print("O número que pensei é maior!")
        elif escolhaComp < numero_selecionado:
            limInf = escolhaComp
            print("O número que pensei é menor!")
        
        escolhaComp = (limInf + limSup) / 2
        tentativas = tentativas + 1
    print("O número escolhido foi", escolhaComp, ", acertou em ", tentativas, "tentativas!!!")



```
